```python
def isCorrectOrder(rules, pages):
    printed = []
    for page in pages:
        if page in rules and len(set(rules[page]).intersection(pages) - set(printed)) > 0:
            #print(requiredPagesPrinted(rules[page],printed), rules[page], printed, page)
            return printed, False
        else:
            printed.append(page)
    return printed, True
```
