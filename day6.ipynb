```python
map = []

with open('Input_Day6.txt', 'r') as file:
    # Read each line in the file
    for line in file:
        map.append(list(line))
```

```python
def findStartingLocation(map):
    for row in range(len(map)):
        for col in range(len(map[row])):
            if(map[row][col] == "^"):
                return (row,col)

currentLocation = findStartingLocation(map)
```

```python
def move(map, nextLocation, direction, pastLocationsAndDirections):
    pastLocations = []
    for t in pastLocationsAndDirections:
        pastLocations.append(t[0])
    if nextLocation not in pastLocations:
        pastLocationsAndDirections.append((nextLocation, direction))
    currentLocation = nextLocation
    while (nextLocation[0]+direction[0] < len(map) and nextLocation[0]+direction[0] >= 0 and nextLocation[1]+direction[1] < len(map[0]) and nextLocation[1]+direction[1] >= 0) and map[nextLocation[0]+direction[0]][nextLocation[1]+direction[1]] == "#":
        direction = turn(direction)
    nextLocation = (currentLocation[0] + direction[0], currentLocation[1] + direction[1])
    
    return pastLocationsAndDirections, nextLocation, direction


def turn(direction):
    goRight = (0,1)
    goDown = (1,0)
    goLeft = (0,-1)
    goUp = (-1,0)

    if direction == goUp:
        return goRight
    if direction == goRight:
        return goDown
    if direction == goDown:
        return goLeft
    if direction == goLeft:
        return goUp
    
def detectLoop(pastLocationsAndDirections, nextLocation, direction):
    return (nextLocation, direction) in pastLocationsAndDirections
```

```python
def lookRightForObstacle(map, position, direction):
    lookDirection = turn(direction)
    nextLocation = (position[0] + lookDirection[0], position[1] + lookDirection[1])
    while nextLocation[0] < len(map) and nextLocation[0] >= 0 and nextLocation[1] < len(map[0]) and nextLocation[1] >= 0:
        if(map[nextLocation[0]][nextLocation[1]] == "#"):
            return True
        nextLocation = (nextLocation[0] + lookDirection[0], nextLocation[1] + lookDirection[1])
    return False

```
